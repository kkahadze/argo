```python
import os
import json
```

```python
prompt = "ბჯოხო"
```

```python
# Cycle through all eaf files in /Users/konstantinekahadze/Desktop/projects/margo/argo/data/iliauni_eafs
for file in os.listdir('/Users/konstantinekahadze/Desktop/projects/margo/argo/data/iliauni_eafs'):
    if file.endswith('.json'):
        with open(os.path.join('/Users/konstantinekahadze/Desktop/projects/margo/argo/data/iliauni_eafs', file), 'r') as f:
            data = json.load(f)
            if data['words']:
                print([item['translation'] for item in data['words'] if item['original'] == prompt])
```
